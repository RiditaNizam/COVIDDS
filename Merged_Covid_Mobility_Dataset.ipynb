G&P 5 Days Average

```python
FiveDayGP = []
FiveDayPP = []

i=5
while(i < len(GP)-1):
  sum5 = GP[i] + GP[i-1] + GP[i-2] + GP[i-3] + GP[i-4]
  avg = sum5 / 5
  FiveDayGP.append(avg)

  i+=5

print("FiveDayGP", len(FiveDayGP))

j=4
while(j < len(PP)-1):
  
  sum5PP = PP[j] + PP[j-1] + PP[j-2] + PP[j-3] + PP[j-4] 

  avgPP = sum5PP / 5
  
  FiveDayPP.append(avgPP)
  j+=5

print("final0", FiveDayPP)
```

```python
FiveDayGP.pop(0)
FiveDayPP = FiveDayPP[:-1]

PPtoArray = np.array(FiveDayPP)
GPtoArray = np.array(FiveDayGP)

print("fiveDPP length", len(FiveDayPP))

plt.plot(GPtoArray, PPtoArray, 'bs')
plt.show
```
